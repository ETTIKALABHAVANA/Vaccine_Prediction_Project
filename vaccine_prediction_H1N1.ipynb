# splitting Data

```python
x = df.iloc[:,:-2]
y = df.iloc[:,-2]
x
```

# Checking Imbalance Data

```python
fig, ax = plt.subplots(2, 1, sharex=True)
n_obs = df2.shape[0]

(df2['h1n1_vaccine']
    .value_counts()
    .div(n_obs)
    .plot.barh(title="H1N1 Vaccine", ax=ax[0])
)
ax[0].set_ylabel("h1n1_vaccine")

(df2['seasonal_vaccine']
    .value_counts()
    .div(n_obs)
    .plot.barh(title="Seasonal Vaccine", ax=ax[1])
)
ax[1].set_ylabel("seasonal_vaccine")

fig.tight_layout()
```

# Balancing Imbalace data

```python
from imblearn.over_sampling import RandomOverSampler
os = RandomOverSampler()
x_train_res,y_train_res = os.fit_sample(x,y)
x_train_res.shape,y_train_res.shape
```

```python
from collections import Counter
print('original dataset shape {}'.format(Counter(y)))
print('Resampled dataset shape{}'.format(Counter(y_train_res)))
```

# Train and Test Model

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_train_res,y_train_res,test_size=0.20,random_state = 42)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,log_loss,roc_auc_score
```

# Extra Tree Algorithm

```python
from sklearn.ensemble import ExtraTreesClassifier
model_ec = ExtraTreesClassifier()
model_ec.fit(x_train,y_train)
ec_pred = model_ec.predict(x_test)
```

```python
print(confusion_matrix(y_test,ec_pred))
print(log_loss(y_test,ec_pred))
print(classification_report(y_test,ec_pred))
```

```python
ec_pred = model_ec.predict_proba(x_test)[:,1]
```

```python
roc_auc_score(y_test,ec_pred)
```

```python
from sklearn.metrics import roc_curve,roc_auc_score
ec_pred = model_ec.predict_proba(x_test)[:,1]
ftr,tpr,thresholds = roc_curve(y_test,ec_pred)
plt.plot(ftr,tpr,lw = 6)
plt.plot([0,1],[0,1],color = "green",linestyle = "dashdot",lw=2)
plt.xlabel('\nFalse Positive Rate(1-Specificity)')
plt.ylabel('\nTrue Positive RAte(Sensitivity)\n')
plt.title(f"seasonal_vaccine ROC_Curve:AUC={roc_auc_score(y_test,ec_pred):4f}\n")
```
